```python
import requests
```

```python
url = 'https://www.taifex.com.tw/cht/3/dailyFXRate'
d = {'queryStartDate':'2018/01/01', 'queryEndDate':'2018/12/31'}
r = requests.post(url, data=d)
print(r.text)
```

```python
import pandas as pd
table = pd.read_html(r.text)
```

```python
df = table[2]
df.head(10)
```

```python
cols = df.loc[0]
df = df.drop(index=[0])
df.columns = cols
df.head(10)
```

```python
df['日期'] = pd.to_datetime(df['日期'])
df.index = df['日期'] 
df.head(10)
```

```python
ds = df[['日期','美元／新台幣']]
ds.columns = ['Date','USD/TWD']
ds['USD/TWD'] = pd.to_numeric(ds['USD/TWD'])
ds.head(10)
```

```python
%pylab inline
ds['USD/TWD'].plot(kind = 'line', figsize=[20, 5])
```

```python
ds['ma5'] = ds['USD/TWD'].rolling(window = 5).mean()
ds['ma10'] = ds['USD/TWD'].rolling(window = 10).mean()
ds['ma21'] = ds['USD/TWD'].rolling(window = 21).mean()
ds.head(25)
```

```python
ds[['USD/TWD','ma5','ma10','ma21']].plot(kind = 'line', figsize=[20, 5])
```
